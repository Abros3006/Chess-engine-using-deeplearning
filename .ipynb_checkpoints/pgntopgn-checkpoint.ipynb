```python
import chess.pgn

def separate_pgn(input_pgn_file, output_directory='output_games/'):
    with open(input_pgn_file) as f:
        pgn = chess.pgn.read_game(f)
        game_counter = 1

        while pgn is not None:
            # Create a new PGN file for each game
            output_file_path = f"{output_directory}game_{game_counter}.pgn"
            
            with open(output_file_path, 'w') as output_file:
                # Write the current game to the new file
                output_file.write(str(pgn))

            print(f"Game {game_counter} saved to {output_file_path}")
            game_counter += 1

            # Read the next game from the input PGN file
            pgn = chess.pgn.read_game(f)

if __name__ == "__main__":
    input_pgn_file = 'C:/Users/athar/Downloads/Morphy/Morphy.pgn'  # Replace with the path to your input PGN file
    output_directory = 'C:/Users/athar/Downloads/Morphy/morphy2/'  # Specify the desired output directory
    
    separate_pgn(input_pgn_file, output_directory)
```
